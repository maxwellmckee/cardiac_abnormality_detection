### Overview: ECG Abnormality Detection with FPCA and B-Splines

I use a combination of Functional Principal Component Analysis (FPCA), B-Spline Regression, and a Random Forest Classifier detect cardiac abnormalities from electrocardiogram (ECG) data. Given the critical nature of timely and accurate detection of heart diseases, this notebook explores the potential of computational methods to supplement traditional medical analysis. The primary goal is to distinguish between normal and abnormal ECG signals by analyzing them as functional data. This involves:

**Data Utilization**: Uses an ECG dataset to model ECG signals, sourced from the UCR Time Series Classification Archive.

**Feature Extraction**: Implementing B-splines and FPCA to effectively capture the key characteristics of ECG signals that indicate normal or abnormal cardiac activity.

**Classification Strategy**: Applying a Random Forest classifier to the extracted features to accurately classify ECG results into normal or abnormal categories, based on their signal shapes and patterns. This approach aims to enhance the precision of cardiac disease detection by integrating advanced data processing and machine learning methodologies with traditional ECG analysis techniques.

## Imports

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from skfda.representation.basis import BSpline
from skfda.representation import FDataGrid
from sklearn.metrics import mean_squared_error

from skfda.representation import FDataGrid
from skfda.preprocessing.dim_reduction.projection import FPCA
```

## Set up data

```python
train_data = pd.read_csv('ecg_train', header=None)
X_train = np.array(train_data)[:,1:]
y_train = np.array(train_data)[:,0]

test_data = pd.read_csv('ecg_test', header=None)
X_test = np.array(test_data)[:,1:]
y_test = np.array(test_data)[:,0]

train_data
```

## Dimensionality Reduction
-------------------------
For the sake of comparing the effectiveness of dimensionality reduction using B-splines to FPCA on the above data, I will reduce them to the same dimensionality: from 96 features to 10 features.

### B-Splines

```python
def bspline_reduction(X_train, X_test, n_knots):
    X_train_reduced = np.zeros((X_train.shape[0], n_knots)) #initialize reduced matrix for X_train
    X_test_reduced = np.zeros((X_test.shape[0], n_knots)) #initialize the reduced matrix for X_test
    basis_objects = [] 
    
    for i in range(X_train.shape[0]):
        knots = np.linspace(np.min(X_train[i]), np.max(X_train[i]), n_knots) #generate equally spaced n_knots
        basis = BSpline(n_basis=n_knots, order=3) #generate a b-spline basis with specified number of basis
        fd = FDataGrid(data_matrix=[X_train[i]], grid_points=np.arange(X_train.shape[1])) #FDataGrid object with X_train[i] as data_matrix and index as grid_points
        fd_basis = fd.to_basis(basis) #fit b-spline to data
        X_train_reduced[i] = fd_basis.coefficients.flatten() #store coefficients in the reduced matrix
        basis_objects.append(basis) # store the basis object

    # apply trained transformation to X_test
    for i in range(X_test.shape[0]):
        fd_test = FDataGrid(data_matrix=[X_test[i]], grid_points=np.arange(X_test.shape[1]))
        fd_test_basis = fd_test.to_basis(basis_objects[i]) # use the basis object obtained from corresponding instance in x
        X_test_reduced[i] = fd_test_basis.coefficients.flatten()

    return X_train_reduced, X_test_reduced

bspline_Xtrain_reduced, bspline_Xtest_reduced = bspline_reduction(X_train, X_test, 10)
bspline_Xtest_reduced.shape
```

### FPCA

```python
def fpca_reduction(X_train, X_test, n_components):
    # Create FDataGrid objects for the entire dataset
    fd_train = FDataGrid(data_matrix=X_train, grid_points=np.arange(X_train.shape[1]))
    fd_test = FDataGrid(data_matrix=X_test, grid_points=np.arange(X_test.shape[1]))
    
    # Apply FPCA
    fpca = FPCA(n_components=n_components)
    X_train_fpca = fpca.fit_transform(fd_train)
    X_test_fpca = fpca.transform(fd_test)

    return X_train_fpca, X_test_fpca

FPCA_Xtrain_reduced, FPCA_Xtest_reduced = fpca_reduction(X_train, X_test, 10) 
FPCA_Xtest_reduced.shape
```

## Classifications with Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix

def train_and_evaluate(X_train, y_train, X_test, y_test):
    clf = RandomForestClassifier()
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    cm = confusion_matrix(y_test, y_pred)
    return cm

cm_bsplines = train_and_evaluate(bspline_Xtrain_reduced, y_train, bspline_Xtest_reduced, y_test)
cm_fpca = train_and_evaluate(FPCA_Xtrain_reduced, y_train, FPCA_Xtest_reduced, y_test)
```

```python
def plot_confusion_matrix(cm, labels=["Abnormal", "Normal"], cmap_colors='Blues', title='Confusion Matrix'):
    df_cm = pd.DataFrame(cm, index=labels, columns=labels)
    plt.figure(figsize=(5,5))
    sns.heatmap(df_cm, annot=True, fmt="d", cmap=cmap_colors, cbar=False, square=True)
    plt.title(title)
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()

plot_confusion_matrix(cm_bsplines, title='B-Splines Random Forest Confusion Matrix')
plot_confusion_matrix(cm_fpca, cmap_colors = 'Greens', title='FPCA Random Forest Confusion Matrix')
```
